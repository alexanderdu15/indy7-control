Trajectory Generation for Indy7 Manipulator

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Simple trajectory


num_points = 2000

t = np.linspace(0, 6*np.pi, num_points)

# 6 by num_points matrix
points = np.zeros((num_points, 6))

for i in range(num_points):
    points[i, :] = np.array([np.cos(t[i]), np.sin(t[i]), np.cos(t[i]), np.sin(t[i]), np.cos(t[i]), np.sin(t[i])])

# convert to degrees
points = points * 180 / np.pi



for i in range(10):
    print(points[i, :])


```

```python

#Save to csv
np.savetxt('../../examples/trajectories/simple_trajectory.csv', points, delimiter=',')
```

```python
# visualize
plt.plot(points[:, 0], points[:, 1])
plt.show()
```
